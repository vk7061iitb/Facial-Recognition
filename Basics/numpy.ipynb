### Integrals and Derivatives


```python
x = np.linspace(1, 10, 100)
y = 1/x**2 * np.sin(x)
dydx = np.gradient(y, x)
y_integral = np.cumsum(y)*(x[1]-x[0])
```

```python
plt.plot(x, y)
plt.plot(x, dydx)
plt.plot(x, y_integral)
```

## Question 1:

let $y = exp(-x/10)sin(x)$. Consider 10000 $x$ intervals in the range $[0, 10]$

1. Plot the function $y$ _vs_ $x$ int the range $[0, 10]$
2. Compute the mean and the standard deviation of $y$ for $x$ in the range $[4, 7]$
3. For $x$ in the range $[4, 7]$, find the value $y_m$ such that $80%$ of $y$ values are less than $y_m$
4. Plot $dy/dx$ vs $x$
5. Find the location where $dy/dx = 0$


```python
def f(x):
    return np.exp(-x/10) * np.sin(x)


N = 10000
x = np.linspace(0, 10, N + 1)
y = f(x)
```

```python
plt.plot(x, y)
```

```python
np.mean(y[(x >= 4)*(x <= 7)])
```

```python
np.std(y[(x >= 4)*(x <= 7)])
```

```python
np.percentile(y[(x >= 4)*(x <= 7)], 80)
```

```python
dydx = np.gradient(y, x)
plt.plot(x, y)
plt.plot(x, dydx)
```

```python
x[1::][dydx[1::]*dydx[:-1:] < 0]
# the expression x[1::][dydx[1::]*dydx[:-1:] < 0] selects elements from the array x based
# on the condition that the product of consecutive elements in dydx is negative.
# The resulting array contains those selected elements.
```

## System of equations
$$ 3x + 2y + z = 4, 5x - 5y + 4z = 3, 6x + z = 0 $$

```python
A = np.array([
    [3, 2, 1],
    [5, -5, 4],
    [6, 0, 1]
])

b = np.array([4, 3, 0])
```

```python
np.linalg.solve(A, b)
```
